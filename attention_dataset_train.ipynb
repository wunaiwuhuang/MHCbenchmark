```python
import pandas as pd
path_attention = '/data1/wuguojia/data/mhc_benchmark/attentionbase'
path_test = '/data1/wuguojia/data/mhc_benchmark/testbase'
#read files
df_1 = pd.read_csv(f'{path_test}/auroc_hla:len2tool_filter.csv')
df_2 = pd.read_excel(f'{path_test}/TEST.xlsx', sheet_name=0)
df_3 = pd.read_excel(f'{path_test}/TEST.xlsx', sheet_name=2)
```

```python
#choose certain data
cols_to_keep = ['allele', 'peptide_length', 'bigmhc', 'capsnetmhc_an', 'mhcflurry_ba', 
                'mhcflurry_ps', 'netmhcpan_ba', 'netmhcpan_el', 'stmhcpan', 'transphla']
length = 9
bind = 'positive'
df_1 = df_1[df_1['peptide_length'] == length][cols_to_keep]
df_2 = df_2[(df_2['antigen_peptide_length']==length)&(df_2['bind_result'] == 'positive')]
df_3.rename(columns={df_3.columns[0]: 'allele'}, inplace=True)
df_1 = pd.merge(df_1, df_3, on='allele', how='left')
print(df_1)
```

```python
#left good samples
#clean none number columns
cols_to_check = df_1.columns[1:]
df_cleaned = df_1.dropna(subset=cols_to_check, how='any').copy()
df_cleaned[cols_to_check] = df_cleaned[cols_to_check].apply(pd.to_numeric, errors='coerce')
df_cleaned = df_cleaned.dropna(subset=cols_to_check, how='any')
df_merged = df_cleaned[['allele', 'peptide_length', str(length)]]
df_merged = df_merged.sort_values(by=str(length), ascending=False)
df_merged = df_merged.drop_duplicates(subset='allele')
#select alleles
hla_a = df_merged.loc[df_merged['allele'].str.contains('HLA-A'), 'allele'].head(1)
hla_b = df_merged.loc[df_merged['allele'].str.contains('HLA-B'), 'allele'].head(1)
hla_c = df_merged.loc[df_merged['allele'].str.contains('HLA-C'), 'allele'].head(1)
hla = pd.concat([hla_a, hla_b,hla_c], ignore_index=True)
print(hla)
```

```python
#generate train_raw and train_use datasets
train_number = 1000
seed = 20030622
df_train_raw = df_2[df_2['hlatype'].isin(hla)]
df_train_use = (df_train_raw
                .groupby('hlatype', group_keys=False)  # 按 hlatype 分组，不保留分组键
                .apply(lambda x: x.sample(n=train_number, random_state=seed)))  # 每组抽取 1000 个样本
import os
train_raw_path = os.path.join(path_attention, 'traindata_raw')
train_use_path = os.path.join(path_attention, 'traindata_use')
os.makedirs(train_raw_path, exist_ok=True)
os.makedirs(train_use_path, exist_ok=True)
def save_subsets(df, folder_path, prefix):
    for (hlatype, length), group in df.groupby(['hlatype', 'antigen_peptide_length']):
        filename = f"{hlatype}_{length}.csv"
        group.to_csv(os.path.join(folder_path, filename), index=False)
save_subsets(df_train_raw, train_raw_path, "train_raw")
save_subsets(df_train_use, train_use_path, "train_use")
```

```python
#generate usable training data
import pickle, importlib

def AA_to_num(peptide):
    values = dict([(x.split('=')[1], x.split('=')[0]) for x in AA])
    peptide = np.array(list(map(lambda x: values[x], peptide)))
    return peptide

def num_to_AA(peptide):
    values = dict([(int(x.split('=')[0]), x.split('=')[1]) for x in AA])
    peptide = list(map(lambda x: values[x], peptide))
    str_peptide = ''
    for aa in peptide:
        str_peptide+=aa
    return str_peptide

def transform_train(train_file):
    df = pd.read_csv(train_file)
    peptides = df['antigen_peptide'].tolist()
    train = np.array([AA_to_num(peptide) for peptide in peptides], dtype='<U20')
    return train.astype(float)

AA = ['0=A','1=R','2=N','3=D','4=C','5=E','6=Q','7=G','8=H','9=I','10=L','11=K','12=M','13=F','14=P','15=S','16=T','17=W','18=Y','19=V','20=B','21=J','22=O','23=U','24=X','25=Z']
```

```python
import shap
import numpy as np
import pandas as pd
import os
path_attention = '/data1/wuguojia/data/mhc_benchmark/attentionbase'
train_raw_path = os.path.join(path_attention, 'traindata_raw')
train_use_path = os.path.join(path_attention, 'traindata_use')

all_files = os.listdir(train_use_path)
train_file = [file for file in all_files if file.endswith('.csv')]
for file in train_file:
    file_name, _ = os.path.splitext(file)
    file_path = os.path.join(train_use_path, file)
    train = transform_train(file_path)
    np.save(os.path.join(train_use_path, f"{file_name}.npy"), train)
    train_summary = shap.kmeans(train, 50)
    with open(os.path.join(train_use_path, f"{file_name}.pkl"), 'wb') as filehandler:
        pickle.dump(train_summary, filehandler)
```
